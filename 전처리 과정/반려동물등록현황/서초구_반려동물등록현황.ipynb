```python
import numpy as np
import pandas as pd
```

```python
file_path = "C:/Users/82108/Desktop/jw/프로젝트/반려동물등록현황전처리전/서울_반려동물_등록현황/서울특별시_서초구_동물등록 현황_20230630.csv"
df_raw = pd.read_csv(file_path, encoding='euc-kr')
# 파일불러오기 
# 인코딩 문제 발생, 현재 파일이 EUC-KR로 인코딩되어 있으므로 encoding='euc-kr'으로 지정
```

```python
# 원본 데이터
df_raw.head()
```

```python
# 원본데이터 복사
seocho = df_raw.copy()
```

```python
seocho['구']="서초구"
seocho.head()
```

```python
# 반려동물총계 컬럼 만들기
seocho['반려동물총계'] = seocho['내장형무선인식(RFID)'] + seocho['외장형무선인식(RFID)'] + seocho['인식표']

seocho.head()
```

```python
# 필요한 컬럼만 남기기
seocho = seocho[['구','법정동','동물소유자수','반려동물총계']]

# 컬럼명 변경하기
seocho.rename(columns={'구':'Gu',
                        '법정동':'Dong',
                        '동물소유자수':'Pet_Owners',
                        '반려동물총계':'Total_Registrations'},
                  inplace=True)

seocho
```

```python
# 결측치 제거
seocho.dropna(inplace=True)
seocho
```

```python
# 파일 저장하기
seocho.to_csv("./seocho_지우_v1.csv")
```
