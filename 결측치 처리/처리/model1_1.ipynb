```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
# 선형 회귀분석을 하는 경우에는 위 클래스를 사용

#from sklearn.model_selection import cross_val_score, KFold
#import statsmodels.api as sm
```

```python
model1_test_x = pd.read_csv('./model1_variable_data/model1_test_x.csv')
model1_test_y = pd.read_csv('./model1_variable_data/model1_test_y.csv')
model1_train_x = pd.read_csv('./model1_variable_data/model1_train_x.csv')
model1_train_y = pd.read_csv('./model1_variable_data/model1_train_y.csv')
```

```python
model1_test_x = model1_test_x.drop(['Unnamed: 0'],axis=1)
print(model1_test_x.head(1))

model1_test_y = model1_test_y.drop(['Unnamed: 0'],axis=1)
print(model1_test_y.head(1))

model1_train_x = model1_train_x.drop(['Unnamed: 0'],axis=1)
print(model1_train_x.head(1))

model1_train_y = model1_train_y.drop(['Unnamed: 0'],axis=1)
print(model1_train_y.head(1))
```

모델 1
- x: Total_Registrations, Population
- y: Pet_Owners

```python
model = LinearRegression()
# 이 모델은 입력 데이터와 목표 변수 간의 선형 관계를 모델링
# 새로운 데이터에 대한 예측 수행

x = model1_train_x
y = model1_train_y

result = model.fit(x,y)
# 가중치 값을 추정
# 상수항 결합을 자동으로 해주므로 사용자가 직접 상수항 결합을 할 필요 없음
# >> 모델은 최적의 가중치(계수)와 편향성(상수항)을 추정 
# >> 가중치와 편향을 추정하는 과정에서 잔차 최소화
# 생성된 모델을 입력 데이터(x)와 목표변수(y)에 대해 학습시키는 과정

predict = result.predict(model1_test_x)
predict = pd.DataFrame(predict)

# 실제 소유자 수와 추정된 소유자 수 비교

accuracy_list = []
for i in range(len(predict)):
        print(i)
        print(f'실제 소유자 수 : {model1_test_y.iloc[i,0]}')
        print(f'추정된 소유자 수: {predict.iloc[i,0]:.2f}')
        print(f'예측의 정확도: {100*(1- abs(predict.iloc[i,0] - model1_test_y.iloc[i,0])/model1_test_y.iloc[i,0]):.2f}%')
        accuracy = 100*(1- abs(predict.iloc[i,0] - model1_test_y.iloc[i,0])/model1_test_y.iloc[i,0])
        accuracy_list.append(accuracy)
        
print('-'*30)
total_accuracy = sum(accuracy_list)/len(accuracy_list)

print(f'총 정확도: {total_accuracy:.2f}%')
```
